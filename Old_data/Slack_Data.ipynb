```python
import pandas as pd
import numpy as np
import xlrd.xldate as xd
import matplotlib.pyplot as plt
import matplotlib.ticker
import matplotlib.dates
```

```python
PH_Slack_Analytics = pd.read_excel("PH_Slack_Analytics.xlsx", header=0)

# 2016 Slack Data (starting from June 1st, 2016)
# Total member growth
plt.xlabel('Dates')
plt.ylabel('Members')
plt.gca().set_color_cycle(['red', 'green', 'blue', 'yellow'])

#count starting from June 1st

for i, row in PH_Slack_Analytics[119:].iterrows():
    row['Date'] = row['Date'].to_pydatetime()

plt.plot(PH_Slack_Analytics[119:]['Date'], PH_Slack_Analytics[119:]['Total Users'])
plt.plot(PH_Slack_Analytics[119:]['Date'], PH_Slack_Analytics[119:]['Daily Active Users'])
plt.plot(PH_Slack_Analytics[119:]['Date'], PH_Slack_Analytics[119:]['Daily Users Posting Messages'])

plt.legend(['Total Users', 'Daily Active Users', 'Daily Users Posting Messages'], loc='upper left')

plt.gca().xaxis.set_major_locator(matplotlib.dates.MonthLocator())
plt.gca().xaxis.set_major_formatter(matplotlib.dates.DateFormatter("%Y-%m"))
plt.gcf().autofmt_xdate()

# Get current size
fig_size = plt.rcParams["figure.figsize"]
# Set figure width to 12 and height to 9
fig_size[0] = 12
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size

plt.show()
```

```python
# Message distribution in 2017 Spring semester
# count starting from January 9th

public_msg = 0
private_msg = 0
DMs = 0

fig, (pie1, pie2) = plt.subplots()

for i, row in PH_Slack_Analytics[339:454].iterrows():
    public_msg += row['Messages in Public Channels']
    private_msg += row['Messages in Private Channels']
    DMs += row['Messages in DMs']
    
"""print('Total number of public messages: ', public_msg)
print('Total number of private messages: ', private_msg)
print('Total number of DMs: ', DMs)
print('Total number of messages sent on PH Slack: ', public_msg+private_msg+DMs)
"""
labels = 'Public channels', 'Private channels', 'DMs'
slices = [public_msg, private_msg, DMs]
colors = ['#e8f0ed','#d6ede6', '#aedcce']

plt.title("Message distribution in Spring 2017")
plt.pie(x=slices, labels=labels, colors=colors, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```
